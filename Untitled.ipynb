```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import geopandas as gpd

# pandas 결과값의 표현 범위 소수점 2자리수로 변경
pd.options.display.float_format = '{:.3f}'.format
pd.set_option('display.max_columns', 150)
pd.set_option('display.width', 10000)
```

# Data load

```python
dev = pd.read_csv('./data/indicator/fs_dec.csv').drop('const', axis=1)
cul = pd.read_csv('./data/indicator/fs_cul.csv')
geo = pd.read_csv('./data/indicator/fs_geo.csv')
cluster = pd.read_csv('./data/cluster/cluster_csv.csv')
```

```python
dev = pd.merge(dev, cluster[['상권_코드', 'cluster_y']], on='상권_코드').set_index('상권_코드')
cul = pd.merge(cul, cluster[['상권_코드', 'cluster_y']], on='상권_코드').set_index('상권_코드')
geo = pd.merge(geo, cluster[['상권_코드', 'cluster_y']], on='상권_코드').set_index('상권_코드')
geo.head()
```
